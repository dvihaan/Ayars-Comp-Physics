```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ya(x):
    return x**4*np.exp(-2*x)

def yb(x):
    return (x**2*np.exp(-x)*np.sin(x**2))**2
```

```python
x = np.linspace(0, 5, 500)

A_values = ya(x)
B_values = yb(x)

graph_A = plt.plot(x,A_values,'r')
graph_B = plt.plot(x,B_values,'b')
plt.title('y vs x')
plt.xlabel('X')
plt.ylabel('f(X)')
```
